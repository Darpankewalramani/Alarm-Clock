```python
from tkinter import *
from tkinter import messagebox
import time
import datetime
import threading
from pygame import mixer
from tkinter import filedialog

root= Tk()
root.title("Alarm")
root.geometry("900x750")

mixer.init()

def th():
    t1 = threading.Thread(target=a, args=())
    t1.start()


def a():
    a = hr.get()
    if a == "":
        msg = messagebox.showerror('Invalid data','Please enter valid time')
    else:
        alarm_time = datetime.datetime.strptime(a, "%H:%M:%S")
        message = amsg.get()
        weekdays = [mon_var.get(), tue_var.get(), wed_var.get(), thu_var.get(), fri_var.get(), sat_var.get(), sun_var.get()]
        snooze_time = 5  # snooze time in minutes
        snooze_count = 0
        
        while True:
            now = datetime.datetime.now()
            if weekdays[now.weekday()] and now.time() >= alarm_time.time():
                while True:
                    mixer.music.load('tone2.mp3')
                    mixer.music.load('tone3.mp3')
                    mixer.music.load('tone4.mp3')
                    mixer.music.play()
                    snooze_response = messagebox.askyesno("please wake up",f"{message}\n\nSnooze for {snooze_time} minutes?")
                    if snooze_response == True:
                        mixer.music.stop()
                        snooze_count += 1
                        if snooze_count == 1:
                            snooze_until = datetime.datetime.now() + datetime.timedelta(minutes=snooze_time)
                        else:
                            snooze_until = snooze_until + datetime.timedelta(minutes=snooze_time)
                        while datetime.datetime.now() < snooze_until:
                            time.sleep(1)
                        break
                        
                    else:
                        mixer.music.stop()
                        break
            time.sleep(1)



header =Frame(root)
header.place(x=5,y=5)

head =Label(root,text="ALARM CLOCK",font=('comic sans',20))
head.pack(fill=X)

panel = Frame(root)
panel.place(x=5,y=70)

alpp = PhotoImage(file='clock.png')

alp = Label(panel,image=alpp)
alp.grid(rowspan=8,column=0)

atime = Label(panel,text="Alarm Time \n(Hr:Min:Sec)",font=('comic sans',25))
atime.grid(row=0,column=1,padx=15,pady=7)

hr = Entry(panel,font=('comic sans',20),width=10, highlightthickness=2)
hr.grid(row=0,column=2,padx=10,pady=5)



amessage = Label(panel,text="Message",font=('comic sans',20))
amessage.grid(row=1,column=1,columnspan=2,padx=10,pady=5)

amsg = Entry(panel,font=('comic sans',15),width=30, highlightthickness=2)
amsg.grid(row=2,column=1,columnspan=2,padx=10,pady=10)


    
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
row = 3
for i, day in enumerate(weekdays):
    var = BooleanVar()
    Checkbutton(panel, text=day, variable=var).grid(row=row+i//4, column=i%4+1)
    if day == "Monday":
        mon_var = var
    elif day == "Tuesday":
        tue_var = var
    elif day == "Wednesday":
        wed_var = var
    elif day == "Thursday":
        thu_var = var
    elif day == "Friday":
        fri_var = var
    elif day == "Saturday":
        sat_var = var
    elif day == "Sunday":
        sun_var = var

        ringtone_label = Label(panel, text="Ring Tone", font=('comic sans', 20))
ringtone_label.grid(row=35, column=1, padx=4, pady=5)

def browse_file():
    file_path = filedialog.askopenfilename()
    ringtone.set(file_path)

ringtone = StringVar()
ringtone.set('tone2.mp3')
ringtone_entry = Entry(panel, textvariable=ringtone, font=('comic sans', 15), width=20, highlightthickness=2)
ringtone_entry.grid(row=35, column=2, padx=10, pady=5)
browse_button = Button(panel, text="Browse", font=('comic sans', 15), command=browse_file)
browse_button.grid(row=35, column=3, padx=10, pady=5)




start = Button(panel,text="Start alarm",font=('comic sans',20),command=th)

start.grid(row=row+40,column=1,columnspan=3,padx=10,pady=5)




root.mainloop()
```
